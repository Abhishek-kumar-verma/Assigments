```python
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
y=Boston['MEDV']
X=Boston.drop(columns="MEDV",axis=1)
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size = 0.25,random_state=355)
```

```python
regression = LinearRegression()

regression.fit(x_train,y_train)
```

```python
regression.score(x_train,y_train)
```

```python
adj_r2(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
adj_r2(x_test,y_test)
```

```python
lassocv=LassoCV(cv=10,max_iter=10000,normalize=True)
lassocv.fit(x_train,y_train)
alpha=lassocv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
alphas=np.random.uniform(low=0.0, high=10,size=(50,))
ridgecv=RidgeCV(alphas=alphas,cv=10,normalize=True)
ridgecv.fit(x_train,y_train)
alpha=ridgecv.alpha_
alpha
```

```python
ridge_reg=Ridge(alpha)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
elastic_cv=ElasticNetCV(alphas=None ,cv=10)
elastic_cv.fit(x_train,y_train)
```

```python
alpha=elastic_cv.alpha_
alpha
```

```python
elastic_cv.l1_ratio
```

```python
elastic_reg=ElasticNet(alpha=alpha,l1_ratio=0.5)
elastic_reg.fit(x_train,y_train)
```

```python
elastic_reg.score(x_test,y_test)
```
